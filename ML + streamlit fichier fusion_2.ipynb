```python
import streamlit as st
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import requests
import json
```

```python

st.title('Bienvenue sur Netflox')
st.image('https://media1.tenor.com/m/Hh5lO8fCAAIAAAAC/netflix.gif', caption='Netflix')




df = ex_Ml.copy() #remplacement  du fichier github par dataset TMDB+ IMDB

movie_name="title"

st.dataframe(df.head())

choix = st.selectbox(
    "Choisi un film et je te proposerai des recommandation qui te plairont :-)",
    df[movie_name])




st.write("Ton choix :", choix)

try:
    infosDuFilm = json.loads(requests.get(f"http://www.omdbapi.com/?apikey=548c959e&t={choix}").text)
    image = infosDuFilm["Poster"]
    st.image(image)
except:
    st.image("https://media.tenor.com/5SSLJ7zpa7EAAAAM/asd.gif")

nn = NearestNeighbors(n_neighbors=6, metric="euclidean")

X = df.drop(columns=movie_name)
#équivalent_de : 
X = df.drop(movie_name, axis=1)
y = df[movie_name]

#Clustering, donc pas de y
nn.fit(X)

#On récupère les caractéristiques du film choisie par l'utilisateur
numeroDelaLigne = df[df[movie_name] == choix].index[0]

#On récupère toutes les colonnes features du film ayant l'index du choix
featuresAPredire = X.iloc[numeroDelaLigne, :]

#featuresAPredire = df.iloc[numeroDelaLigne, :].drop(columns="movie_name")

#st.write(featuresAPredire)

distances, index = nn.kneighbors([featuresAPredire])

listeDesRecommandations = df[movie_name][index[0]]

st.write(distances)
st.write(index)
st.write(listeDesRecommandations.values)

st.write(listeDesRecommandations.values[1:])
```
